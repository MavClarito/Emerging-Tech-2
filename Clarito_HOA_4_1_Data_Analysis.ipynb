**Titanic_train plotting tests.**

```python
menDf = titanicDF1[(titanicDF1.Sex== 'male')]
womenDf = titanicDF1[(titanicDF1.Sex== 'female')]
```

```python
menMean = menDf[["Age", "Pclass"]].mean(axis=1) #Gets the mean of the two variables
plt.scatter(menMean, menDf["Fare"]) #Plots the two variables specifically the mean of two and the fare of the Men
plt.show()
%matplotlib inline
```

```python
womenMean = womenDf[["Age", "Pclass"]].mean(axis=1) #Gets the mean of the two variables
plt.scatter(womenMean, womenDf["Fare"]) #Plots the two variables specifically the mean of two and the fare of the women
plt.show()
%matplotlib inline
```
