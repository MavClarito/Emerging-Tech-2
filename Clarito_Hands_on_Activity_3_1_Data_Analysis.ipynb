```python
# On this cell is where we import modules and load the data set that we downloaded (name_gender_dataset.csv)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

heartdiseaseFile = '/content/behavioral_risk_factor_data_heart_disease_stroke_prevention.csv'
heartFrame = pd.read_csv(heartdiseaseFile, sep = ',')
```

```python
#Displays the first 5 names on the data set.
heartFrame.head()
```

```python
#Displays the last 5 names on the data set.
heartFrame.tail()
```

```python
#Brief description that consists of minimum, maximum Quartile, standard deviation, mean and count of the datas in the data set.
heartFrame.describe()
```

```python
#Creates a frame for two genders.
lowConfidenceDf = heartFrame[heartFrame['confidence_limit_low'] < 0.5]
highConfidenceDf = heartFrame[heartFrame['confidence_limit_high'] > 0.8]
```

```python
# Create a scatter plot with confidence limit low and high for the entire dataset
plt.scatter(heartFrame["confidence_limit_low"], heartFrame["confidence_limit_high"])
plt.xlabel('Confidence Limit Low')
plt.ylabel('Confidence Limit High')
plt.title('Scatter plot of Confidence Limits')
plt.grid(True)
plt.show()
```

```python
heartFrame.corr(method='pearson')
```

```python
wcorr = lowConfidenceDf.corr()
sns.heatmap(wcorr)
```

```python
mcorr = highConfidenceDf.corr()
sns.heatmap(mcorr)
```
